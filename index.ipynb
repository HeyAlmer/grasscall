```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import mpld3
from mpld3 import plugins
mpld3.enable_notebook()

#data
x = [0.0, 2, 3.7, 4.5, 4.75, 5.5,  5, 3, 5.5, 6.75,  7, 7, 7.25, 7.5, 7.5, 10, 8.5, 9.5, 10, 10.5, 10.75, 11, 16]
y = [1.5, 2, 2.5, 6.0, 3.00, 6.5, 10, 9, 11, 7, 4, 3, 5.00, 7.0, 9.5, 8,  8, 3, 4, 2.5, 5, 2.5, 2 ] 
z = [4, 7, 5, 3, 5, 7, 3, 6, 25, 5, 3, 7, 6, 2, 5, 4, 10, 3, 7, 5, 3, 7, 3]    

fig, ax = plt.subplots()
fig.set_size_inches(12, 6)
ax.grid(True, alpha=0.3)
   
scatter = ax.scatter(x, y, color='g',s =  10 * np.array(z), alpha=0.3)
ax.plot(x, y, linestyle="-", color="g", alpha=0.8)

ax.set_xlim([0, 16])
ax.set_ylim([0, 16])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Grass Call', size=16)
labels = z
tooltip = mpld3.plugins.PointLabelTooltip(scatter, labels=labels)
plugins.connect(fig, tooltip)

# plt.savefig('gc_big.png')
```
